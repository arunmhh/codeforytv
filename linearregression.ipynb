```r
inhp <- read.csv("income.data.csv")
print(inhp)
```

```r
ggscatterstats(data = inhp,
x = income,
y = happiness,
title = "Happiness vs. Income",
xlab = "Income",
ylab = "Happiness",
type = "parametric",
p.adjust.method = "bonferroni")
```

```r
m <- lm(income ~ happiness, data = inhp)
summary(m)
```

```r
check_model(m)
```

```r
plot_model(m,type = "eff", terms = "happiness", show.data = T)
```

```r
plot_model(m,show.values = T, terms = "happiness")
```
