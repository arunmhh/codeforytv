```python
import GEOparse
import pandas as pd
import seaborn as sns
```

```python
gse = GEOparse.get_GEO(geo= "GSE1563",destdir= "./GEO")
```

```python
gse_matadata = pd.DataFrame(list(gse.metadata.items()),
                  columns = ["Key","Value"])

print(gse_matadata)
```

```python
pivoted_control_samples = gse.pivot_samples('VALUE')
pivoted_control_samples.head()
```

```python
pivoted_control_samples.hist()
sns.despine(offset=10, trim=True)
```

```python
pivoted_control_samples_average = pivoted_control_samples.median(axis=1)
print("Number of probes before filtering: ", len(pivoted_control_samples_average))
```

```python
expression_threshold = pivoted_control_samples_average.quantile(0.25)
```

```python
expressed_probes = pivoted_control_samples_average[pivoted_control_samples_average >= expression_threshold].index.tolist()
print("Number of probes above threshold: ", len(expressed_probes))
```

```python
samples = gse.pivot_samples("VALUE").loc[expressed_probes]
print(samples)
```

```python
meta = []

for gsm_name, gsm in gse.gsms.items():
    title = gsm.metadata.get("title", [""])[0]  # sample title
    meta.append({
        "gsm": gsm_name,
        "title": title
    })

# Create DataFrame
pheno = pd.DataFrame(meta)
print("Pheno shape:", pheno.shape)
print(pheno.head(10))
print(pheno)
```
